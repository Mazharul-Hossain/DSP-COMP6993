```python
from libpy import NS_dp
```

```python
df = NS_dp.clean_Ames_Housing(AmesHousing)
```

```python
from libpy import FS
```

```python
from sklearn.utils import shuffle
```

```python
X_train, y_train, X_test, y_test = FS.feature_select(df, k=100)
```

```python
X_train.sample(10)
```

```python
from sklearn import linear_model
```

```python
from libpy import CV
```

```python
from sklearn.svm import SVR
```

```python
score, svr_rbf = CV.validate_model( SVR(kernel='rbf', C=1e3, gamma=0.1), X_train, y_train )

score
```

```python
score, svr_lin = CV.validate_model( SVR(kernel='linear', C=1e3), X_train, y_train )

score
```

```python
score, svr_poly = CV.validate_model( SVR(kernel='poly', gamma='auto', C=1e3, degree=2), X_train, y_train )

score
```

```python
predictions = {}
```

```python
y_pred = svr_rbf.predict( X_test ).flatten()

prediction = []
prediction.append( y_test.values.flatten() )
prediction.append(y_pred)
predictions["svr_rbf"] = prediction 

CV.plot_predict_error(y_test, y_pred)
```

```python
y_pred = svr_lin.predict( X_test ).flatten()

prediction = []
prediction.append( y_test.values.flatten() )
prediction.append(y_pred)
predictions["svr_lin"] = prediction 

CV.plot_predict_error(y_test, y_pred)
```

```python
y_pred = svr_poly.predict( X_test ).flatten()

prediction = []
prediction.append( y_test.values.flatten() )
prediction.append(y_pred)
predictions["svr_poly"] = prediction 

CV.plot_predict_error(y_test, y_pred)
```

```python
bf = CV.pack_data(predictions)
```
