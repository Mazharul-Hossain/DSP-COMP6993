```python
X_train, y_train, X_test, y_test = FS.feature_select(df, k=25)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
columns = X_train.columns
```

```python
 # Don't cheat - fit only on training data
scaler.fit(X_train, y_train)  
X_train = scaler.transform(X_train)
```

```python
X_train = pandas.DataFrame( X_train, columns=columns )
```

```python
X_train.sample(10)
```

```python
# apply same transformation to test data
X_test = scaler.transform(X_test)
```

```python
from sklearn import linear_model
```

```python
from libpy import CV
```

```python
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import RandomForestRegressor

from sklearn.neighbors import KNeighborsRegressor

from sklearn.multioutput import MultiOutputRegressor

from sklearn.neural_network import MLPRegressor

from sklearn import linear_model
from sklearn.linear_model import RidgeCV

from sklearn.svm import SVR
```

```python
# Fit estimators
ESTIMATORS = {
    "Extra trees": ExtraTreesRegressor(n_estimators=100, max_features=10,
                                       random_state=0),
    
    "Random Forest": RandomForestRegressor(random_state=0, n_estimators=100),
#     "Multi Output" : MultiOutputRegressor(RandomForestRegressor(n_estimators=100, max_depth=30, random_state=0)),
    
#     "Neural Network" : MLPRegressor( early_stopping=True, warm_start=False),
    
    "K-nn": KNeighborsRegressor(),
    
    "Linear regression": linear_model.LinearRegression(),
    
    "Ridge CV": RidgeCV(),
    "Ridge" : linear_model.Ridge (alpha = .5),
    
    "Lasso" : linear_model.Lasso(alpha = 0.15, tol=0.001),
    
    "svr_rbf" : SVR(kernel='rbf', C=1e3, gamma=0.1),
    "svr_lin" : SVR(kernel='linear', C=1e3),
    "svr_poly" : SVR(kernel='poly', gamma='auto', C=1e3, degree=2)
}
```

```python
predictions = dict()
```

```python
for name, estimator in ESTIMATORS.items():
    score, estimator = CV.validate_model( estimator, X_train, y_train )
    
    print(score, estimator)
    
    y_pred = estimator.predict( X_test ).flatten()

    prediction = []
    prediction.append( y_test.values.flatten() )
    prediction.append(y_pred)
    predictions[name] = prediction 

    CV.plot_predict_error(y_test, y_pred, name)
```

```python
bf = CV.pack_data(predictions)

bf.sample(20)
```
